## 트리기반분류모델 (Tree-Based Classification Model)
트리 기반 분류 모델은 데이터 분류 작업에 사용되는 기본적인 머신러닝 알고리즘입니다. 이 모델은 데이터 세트를 반복적으로 이진 트리 구조로 나누는 방식으로 작동합니다. 각 노드는 특정 속성에 따라 데이터를 분류하며, 최종 노드(리프 노드)는 클래스 레이블을 결정합니다.

- **데이터 분할을 통해 예측 모델 생성**: 트리 기반 모델은 데이터를 여러 번 분할하여 최종 예측 모델을 만듭니다.
- **이해 및 시각화가 쉬움**: 모델이 결정 트리 형태로 표현되기 때문에 직관적으로 이해하고 시각화하기가 쉽습니다.
- **과적합 문제 발생 가능**: 데이터에 과적합되는 경향이 있어, 신중한 가지치기(pruning)가 필요합니다.

## 랜덤 포레스트 모델 (Random Forest Model)
랜덤 포레스트는 여러 개의 결정 트리(decision tree)를 사용하여 예측을 수행하는 앙상블 학습(ensemble learning) 기법입니다. 각 트리는 훈련 데이터의 다른 부분에서 학습되며, 최종 예측은 모든 트리의 예측을 평균하여 결정됩니다.

- **높은 예측 정확도**: 여러 트리의 예측을 평균화하여 최종 예측을 수행하므로 높은 정확도를 보입니다.
- **과적합 방지**: 여러 트리를 사용하여 예측을 수행하기 때문에 단일 트리보다 과적합의 위험이 줄어듭니다.
- **중요 변수 선택 가능**: 각 변수의 중요도를 평가할 수 있습니다.
- **트리의 수와 깊이에 따라 성능 결정**: 사용되는 트리의 수와 깊이가 모델의 성능에 큰 영향을 미칩니다.

<br/>

### 랜덤 포레스트 모델을 이해하기 위한 선행 지식
랜덤 포레스트를 이해하기 위해서는 결정 트리의 기본 원리와 앙상블 학습 기법에 대한 이해가 필요합니다. 랜덤 포레스트는 다수의 약한 학습기(weak learner)를 결합하여 강력한 학습기(strong learner)를 만드는 방식입니다.

<br/><br/>

## 거리기반분류모델 (Distance-Based Classification Model)
거리 기반 분류 모델은 데이터 포인트 간의 거리를 측정하여 분류를 수행하는 모델입니다. 대표적인 예로 k-최근접 이웃 알고리즘(k-Nearest Neighbors, k-NN)이 있습니다. k-NN은 새로운 데이터 포인트의 클래스를 결정하기 위해 가장 가까운 k개의 이웃 데이터 포인트를 고려합니다.

- **간단하고 직관적인 모델**: k-NN 알고리즘은 이해하기 쉽고 구현이 간단합니다.
- **계산 비용이 높을 수 있음**: 데이터 포인트 간의 거리를 계산해야 하므로 큰 데이터셋에서는 계산 비용이 높을 수 있습니다.
- **데이터의 규모와 차원에 민감**: 데이터의 규모가 커지거나 차원이 높아질수록 성능이 저하될 수 있습니다.
- **거리 측정 방식에 따라 성능 달라짐**: 유클리드 거리, 맨하탄 거리 등 다양한 거리 측정 방식을 사용할 수 있으며, 선택한 방식에 따라 모델 성능이 달라질 수 있습니다.

<br/><br/>


## 회귀 모델 (Regression Model)
회귀 모델은 연속적인 값을 예측하기 위해 사용되는 통계적 기법입니다. 회귀 분석은 독립 변수와 종속 변수 간의 관계를 모델링하여 예측을 수행합니다.

### 주요 유형
- **선형 회귀 (Linear Regression)**: 독립 변수와 종속 변수 간의 선형 관계를 모델링합니다.
- **로지스틱 회귀 (Logistic Regression)**: 종속 변수가 범주형인 경우 사용되며, 결과는 확률로 표현됩니다.

<br/><br/>

## 라쏘 모델 (Lasso Model)
라쏘(Lasso, Least Absolute Shrinkage and Selection Operator) 모델은 선형 회귀 모델의 일종으로, L1 정규화를 적용하여 모델의 복잡도를 제어하고 중요하지 않은 변수의 계수를 0으로 만들어 변수 선택을 자동으로 수행합니다.

### 주요 특징
- **모델의 복잡도 제어**: L1 정규화를 통해 모델의 복잡도를 제어합니다.
- **변수 선택 기능**: 중요하지 않은 변수의 계수를 0으로 만들어 자동으로 변수 선택을 수행합니다.

<br/><br/>

## 릿지 모델 (Ridge Model)
릿지(Ridge) 회귀 모델은 선형 회귀 모델에 L2 정규화를 적용하여 과적합을 방지하는 모델입니다. 모든 변수의 계수를 작게 만들어 모델의 복잡도를 줄이는 방식입니다.

### 주요 특징
- **과적합 방지**: L2 정규화를 통해 모델의 복잡도를 줄여 과적합을 방지합니다.
- **모든 변수 사용**: 변수의 계수를 작게 만들지만, 모든 변수를 사용합니다.


<br/><br/>

## Elastic Net 모델 (Elastic Net Model)
Elastic Net 모델은 라쏘와 릿지 회귀의 장점을 결합한 모델로, L1과 L2 정규화를 동시에 사용합니다. 이 모델은 변수 선택과 과적합 방지를 모두 수행합니다.

### 주요 특징
- **혼합 정규화**: L1과 L2 정규화를 동시에 사용하여 모델의 성능을 향상시킵니다.
- **변수 선택과 과적합 방지**: 라쏘의 변수 선택 기능과 릿지의 과적합 방지 기능을 모두 포함합니다.

# L1 정규화 (L1 Regularization)

## 정의
L1 정규화는 머신러닝 모델에서 과적합을 방지하기 위해 가중치의 절대값 합을 최소화하는 기법입니다. 주로 라쏘 회귀(Lasso Regression)에서 사용됩니다.

## 수식
L1 정규화는 다음과 같은 형태의 비용 함수에 추가됩니다: <br/>
#### Cost Function = Loss + λ ∑|β_j|

여기서 λ 는 정규화 강도를 조절하는 하이퍼파라미터이고, β는 모델의 가중치입니다.

## 주요 특징
- **가중치 희소성**: L1 정규화는 중요하지 않은 변수의 가중치를 0으로 만들어 변수 선택을 자동으로 수행합니다.
- **해석 용이성**: 결과 모델이 희소성을 가지므로, 해석이 용이합니다.
- **변수 선택**: 불필요한 변수를 제거하여 모델의 성능을 향상시킵니다.

## 장점
- 불필요한 변수를 제거하여 모델을 단순화합니다.
- 과적합을 방지하여 모델의 일반화 성능을 향상시킵니다.

## 단점
- 데이터가 매우 큰 경우에는 계산 비용이 높아질 수 있습니다.
- 모든 변수의 중요성을 동일하게 고려하지 않기 때문에 중요한 변수가 제거될 가능성도 있습니다.

<br/><br/><br/>

# L2 정규화 (L2 Regularization)

## 정의
L2 정규화는 머신러닝 모델에서 과적합을 방지하기 위해 가중치의 제곱 합을 최소화하는 기법입니다. 주로 릿지 회귀(Ridge Regression)에서 사용됩니다.

## 수식
L2 정규화는 다음과 같은 형태의 비용 함수에 추가됩니다: <br/>
 Loss + λ ∑(β_j^2)
여기서 λ는 정규화 강도를 조절하는 하이퍼파라미터이고, β는 모델의 가중치입니다.

## 주요 특징
- **가중치 축소**: 모든 변수의 가중치를 작게 만들어 과적합을 방지합니다.
- **매끄러운 해결**: L2 정규화는 가중치가 0이 되지 않도록 하여 모든 변수를 모델에 포함시킵니다.
- **안정성**: 높은 다중공선성을 가진 데이터셋에 적합합니다.

## 장점
- 모든 변수를 포함시켜 예측 모델의 안정성을 높입니다.
- 다중공선성이 있는 데이터셋에 잘 작동합니다.
- 과적합을 방지하여 모델의 일반화 성능을 향상시킵니다.

## 단점
- 모든 변수가 모델에 포함되므로 해석이 어려울 수 있습니다.
- 데이터가 희소할 경우 성능이 저하될 수 있습니다.

## 결론
L1과 L2 정규화는 모두 모델의 과적합을 방지하는 데 유용한 기법입니다. L1 정규화는 변수 선택 기능을 제공하며, L2 정규화는 모든 변수를 포함시켜 모델의 안정성을 높입니다. 두 방법의 적절한 선택은 데이터의 특성과 분석 목적에 따라 달라집니다.

# 1. Random Forest

랜덤 포레스트는 앙상블 학습 방법 중 하나로, 여러 개의 의사결정 나무(Decision Tree)를 학습시켜 이를 결합하여 최종 예측을 수행합니다. 개별 나무들이 학습 데이터에 과적합(overfitting)되는 문제를 보완합니다.

### 주요 특징
- **앙상블 학습(Ensemble Learning)**: 여러 개의 의사결정 나무를 결합하여 더 강력한 모델을 만듭니다.
- **배깅(Bagging)**: 학습 데이터의 랜덤 샘플링과 피처 샘플링을 통해 다양한 의사결정 나무를 생성합니다.
- **과적합 방지**: 여러 나무의 결과를 결합함으로써 과적합을 줄입니다.
- **안정성**: 데이터의 노이즈에 민감하지 않으며, 안정적인 예측 성능을 보입니다.

### 작동 방식
1. 학습 데이터를 무작위로 샘플링하여 여러 개의 데이터 셋을 만듭니다.
2. 각 데이터 셋에 대해 의사결정 나무를 학습시킵니다.
3. 분류의 경우 다수결 투표 방식으로, 회귀의 경우 평균을 내어 최종 예측 결과를 도출합니다.

```python
# 참조 코딩
# 범주형 변수를 cols1에 저장.
cols = ['ACCESS', 'ACR', 'BATH', 'BLD', 'COW', 'FESRP', 'FKITP', 'FPARC', 'FSCHP',\
         'FTAXP', 'HHL', 'HHT', 'LANX', 'MAR', 'MV', 'R18', 'R65', 'RAC1P', 'RWAT',\
         'SCH', 'SCHL', 'SEX', 'VALP_B1']

# 범주형 변수의 dtype을 category로 변경
df[cols] = df[cols].astype('category')
df.dtypes

data = df.drop(['VALP_B1'], axis=1)   # 타겟변수를 제외한 입력변수를 data에 저장
target = df['VALP_B1']                # 타겟변수만 target에 저장

# 50:50 데이터 분할
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    data, target, test_size=0.5, random_state=42)   # test_size=0.5임에 주의
```

```python
# Random Forest 모델 (Default 모델, tree depth 제한 없음)
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rf = RandomForestClassifier(n_estimators = 100, random_state=0)
model = rf.fit(X_train, y_train)
pred = model.predict(X_test)  # 학습된 Classifier로 테스트 데이터셋 자료이용해서 타겟변수 예측값 생성
#accuracy = accuracy_score(y_test, pred)

print ("Random Forest Accuracy on training set:{:.5f}".format(model.score(X_train, y_train)))
print ("Random Forest Accuracy on test set:{:.5f}".format(accuracy_score(y_test, pred)))


# Random Forest Accuracy on training set:1.00000
# Random Forest Accuracy on test set:0.83109
```

```python
# Random Forest 모델 (Default 모델, tree depth 제한 없음)
rf = RandomForestClassifier(n_estimators = 100, random_state=0)

# 그리드 서치 실행
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold

# StratifiedKFold의 random_state 옵션값을 특정 숫자(예: 0)로 고정
cross_validation = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
params = {'max_depth': range(10, 41), 'n_estimators': [100, 200]}

# GridSearchCV의 cv=cross_validation 옵션값은 위의 StratifiedKFold의 random_state 옵션값을 적용시켜서
# GridSearchCV를 실행핼 때마다 결과가 항상 동일하게 나오도록 보장
grid_rf = GridSearchCV(rf, param_grid=params, scoring='accuracy', cv=cross_validation,
                       verbose=1, n_jobs=-1)
grid_rf.fit(X_train, y_train)

print("GridSearchCV max accuracy:{:.5f}".format(grid_rf.best_score_))
print("GridSearchCV best parameter:", (grid_rf.best_params_))


# GridSearchCV max accuracy:0.82922
# GridSearchCV best parameter: {'max_depth': 28, 'n_estimators': 200}
```

```python
best_clf = grid_rf.best_estimator_
pred = best_clf.predict(X_test)
print("Accuracy on test set:{:.5f}".format(accuracy_score(y_test, pred)))

from sklearn.metrics import roc_auc_score
ROC_AUC = roc_auc_score(y_test,best_clf.predict_proba(X_test)[:, 1])
print("ROC AUC on test set:{:.5f}".format(ROC_AUC))

# Accuracy on test set:0.83265
# ROC AUC on test set:0.90991
```

# 2. Gradient Boosting

### 개요
그라디언트 부스팅은 여러 약한 학습기(weak learner)를 결합하여 강력한 학습기를 만드는 앙상블 학습 방법입니다. 주로 의사결정 나무를 약한 학습기로 사용합니다. 각 단계에서 모델이 오차를 줄이는 방향으로 예측을 개선해 나갑니다.

### 주요 특징
- **단계적 모델 학습**: 각 단계에서 이전 모델의 오차를 보완하는 새로운 모델을 추가합니다.
- **가중치 업데이트**: 오차가 큰 데이터 포인트에 더 큰 가중치를 부여하여 학습합니다.
- **높은 예측 성능**: 잘 조정된 그라디언트 부스팅 모델은 매우 높은 예측 성능을 보여줍니다.
- **유연성**: 회귀와 분류 문제 모두에 적용 가능합니다.

### 작동 방식
1. 초기 모델을 학습시킵니다.
2. 초기 모델의 예측 결과와 실제 값의 차이를 계산하여 잔여 오차(residual)를 구합니다.
3. 잔여 오차를 예측하는 새로운 모델을 학습시킵니다.
4. 여러 모델의 예측 결과를 결합하여 최종 예측 결과를 도출합니다.

3. Regression
## 개요
회귀(Regression)는 주어진 독립 변수(입력 변수)와 종속 변수(목표 변수) 간의 관계를 모델링하여 새로운 데이터에 대해 종속 변수를 예측하는 통계적 방법입니다. 주로 연속형 데이터를 예측하는 데 사용됩니다.

## 주요 특징
- **목표 변수**: 연속형 데이터. 예를 들어, 집 값, 온도, 매출 등이 회귀 분석의 목표 변수가 될 수 있습니다.
- **독립 변수**: 목표 변수를 설명하는 데 사용되는 변수들. 여러 개의 독립 변수가 있을 수 있습니다.
- **선형 및 비선형 회귀**: 독립 변수와 종속 변수 간의 관계를 직선으로 모델링하는 경우 선형 회귀, 곡선으로 모델링하는 경우 비선형 회귀라고 합니다.

## 종류
- **선형 회귀(Linear Regression)**: 독립 변수와 종속 변수 간의 선형 관계를 가정합니다.
- **다중 회귀(Multiple Regression)**: 여러 독립 변수를 사용하는 선형 회귀입니다.
- **다항 회귀(Polynomial Regression)**: 독립 변수와 종속 변수 간의 비선형 관계를 다항식 형태로 모델링합니다.
- **리지 회귀(Ridge Regression)**: 회귀 계수에 제약을 추가하여 과적합을 방지하는 방법입니다.
- **라쏘 회귀(Lasso Regression)**: 회귀 계수의 절대값 합이 일정 이하가 되도록 제약을 가하는 방법입니다.

## 작동 방식
1. **모델 선택**: 적절한 회귀 모델을 선택합니다 (예: 선형 회귀, 다항 회귀).
2. **모델 학습**: 주어진 데이터로 모델의 계수를 학습합니다. 보통 최소 제곱법(Least Squares Method)을 사용하여 오차의 제곱합을 최소화합니다.
3. **모델 평가**: 학습된 모델의 성능을 평가합니다. 주로 결정계수(R-squared), 평균 제곱 오차(MSE) 등의 지표를 사용합니다.
4. **예측**: 새로운 데이터에 대해 학습된 모델을 사용하여 종속 변수를 예측합니다.
